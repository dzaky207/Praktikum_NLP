```python
# Import Library
import numpy as np
import pandas as pd
import re
import string
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
```

```python
df = pd.read_csv('data.csv')
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
text = df['content']
def hilang_angka_tandaBaca(text):
    clean = re.sub(r'[^a-zA-Z\s]', '', text)
    return clean
df['content']=text.apply(hilang_angka_tandaBaca)
print(df['content'])
```

```python
tfidf = TfidfVectorizer(min_df=5)
tfidf_matrix = tfidf.fit_transform(df['content'])
print(tfidf_matrix)
```

(0, 591) adalah representasi dari indeks dokumen dan indeks kata dalam matriks TF-IDF.

0 berarti dokumen ini merupakan dokumen pertama dalam dataset.

591 mengindikasikan bahwa ini adalah kata ke-591 (indeks dimulai dari 0) dalam matriks TF-IDF.

0.14043768002372942 adalah bobot TF-IDF dari kata dengan indeks 591 dalam dokumen pertama. Ini adalah nilai bobot yang menunjukkan pentingnya kata tersebut dalam dokumen pertama. Semakin tinggi bobotnya, semakin penting kata tersebut dalam konteks dokumen tersebut.

```python
print(f'Jumlah fiturnya adalah {len(tfidf.get_feature_names_out())}.')
print(tfidf.get_feature_names_out())
```

```python
final_data = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
final_data
```
